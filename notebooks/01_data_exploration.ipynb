## Exploring the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("../data/atp_tennis.csv")
```

```python
# Quick overview
display(df.head())
display(df.tail())
```

```python
# Match frequency by surface type
surface_counts = df['Surface'].value_counts()
print("Match frequency by surface:")
print(surface_counts)
```

```python
plt.figure(figsize=(6,4))
colors = ['darkblue', 'orange', 'green', 'limegreen']
plt.bar(surface_counts.index, surface_counts.values, color=colors)
plt.xlabel("Surface Type")
plt.ylabel("Frequency")
plt.title("Number of Matches by Surface Type")
plt.show()
```

## Analyzing Top Winners in Grand Slam Tournaments by Surface

### Hard Courts

```python
hard_surface_filter = ((df['Tournament'] == 'Australian Open') | (df['Tournament'] == 'US Open')) & (df['Round'] == 'The Final')
df_final = df[hard_surface_filter]

hard_winners = df_final['Winner'].value_counts()

print("Top winners on Hard Courts:")
print(hard_winners)
```

```python
hard_winners_name = hard_winners.index
hard_winners_freq = hard_winners.values

plt.barh(hard_winners_name, hard_winners_freq, color='darkblue')
plt.xlabel("Frequency")
plt.ylabel("Players")
plt.title("Top Winners on Hard Courts")
plt.show()
```

### Clay Courts

```python
clay_surface_filter = (df['Tournament'] == 'French Open') & (df['Round'] == 'The Final')
df_final = df[clay_surface_filter]

clay_winners = df_final['Winner'].value_counts()

print("Top winners on Clay Courts:")
print(clay_winners)
```

```python
clay_winners_name = clay_winners.index
clay_winners_freq = clay_winners.values

plt.barh(clay_winners_name, clay_winners_freq, color='orange')
plt.xlabel("Frequency")
plt.ylabel("Players")
plt.title("Top Winners on Clay Courts")
plt.show()
```

### Grass Courts

```python
grass_surface_filter = (df['Tournament'] == 'Wimbledon') & (df['Round'] == 'The Final')
df_final = df[grass_surface_filter]

grass_winners = df_final['Winner'].value_counts()

print("Top winners on Grass Courts:")
print(grass_winners)
```

```python
grass_winners_name = grass_winners.index
grass_winners_freq = grass_winners.values

plt.barh(grass_winners_name, grass_winners_freq, color='green')
plt.xlabel("Frequency")
plt.ylabel("Players")
plt.title("Top Winners on Grass Courts")
plt.show()
```
